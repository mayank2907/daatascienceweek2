```python
# Task 1: Basic Operations
num1 = int(input("Enter the first integer: "))
num2 = int(input("Enter the second integer: "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "Undefined (division by zero)"
modulus = num1 % num2 if num2 != 0 else "Undefined (modulus by zero)"
exponentiation = num1 ** num2

print(f"\nResults of arithmetic operations between {num1} and {num2}:")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {num1} / {num2} = {division}")
print(f"Modulus: {num1} % {num2} = {modulus}")
print(f"Exponentiation: {num1} ** {num2} = {exponentiation}")
```

```python
# Task 2: Working with Lists and Arrays

import numpy as np

numbers = [12, 45, 23, 67, 89, 34, 22, 90, 11, 56]

print(f"Original list: {numbers}")
print(f"Length of the list: {len(numbers)}")
print(f"Maximum value: {max(numbers)}")
print(f"Minimum value: {min(numbers)}")

numbers.append(77)  # Adding 77
removed_element = numbers.pop(3)  # Removing the element at index 3
print(f"List after adding 77 and removing element at index 3 ({removed_element}): {numbers}")

ascending = sorted(numbers)
descending = sorted(numbers, reverse=True)
print(f"Sorted in ascending order: {ascending}")
print(f"Sorted in descending order: {descending}")

array = np.array(numbers)
mean = np.mean(array)
median = np.median(array)
std_dev = np.std(array)

print(f"\nNumPy Array: {array}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

```python
# Task 3: Dictionaries and Sets
student = {
    "name": "Mayank Tyagi",
    "age": 21,
    "course": "Data Science",
    "marks": 88
}

print("Student Information:")
for key, value in student.items():
    print(f"{key}: {value}")

student["grade"] = "A"
print("\nUpdated Student Dictionary with Grade:")
for key, value in student.items():
    print(f"{key}: {value}")

courses = {"Python", "Data Science", "AI", "Python"}
print(f"\nUnique Courses Set: {courses}")

set_a = {"Python", "AI", "ML", "SQL"}
set_b = {"AI", "Data Science", "SQL", "Excel"}

union_set = set_a.union(set_b)
intersection_set = set_a.intersection(set_b)
difference_set = set_a.difference(set_b)

print("\nSet Operations:")
print(f"Set A: {set_a}")
print(f"Set B: {set_b}")
print(f"Union: {union_set}")
print(f"Intersection: {intersection_set}")
print(f"Difference (A - B): {difference_set}")
```
